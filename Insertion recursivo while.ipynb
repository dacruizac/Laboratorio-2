```python
from __future__ import print_function
from itertools import permutations
import numpy as ny
import pandas as pn
import matplotlib.pyplot as plt
```

```python
def insertion_sort(A):
    key=0
    C=[]
    for p in range(len(A)):
        C.append(A[p])
    steps=0
    for i in range(1,len(C)):
        key=C[i]
        j=i-1
        while(j>=0 and key<C[j]):
            C[j+1]=C[j]
            j=j-1
            steps=steps+1
        steps=steps+1
        C[j+1]=key
    return steps
```

```python
def perm(A,P,i,n):
    if (i==n):
        P.append(insertion_sort(A));
    else:
        for j in range(i,n):
            A[j],A[i]=A[i],A[j]
            perm(A,P,i+1,n)
            A[j],A[i]=A[i],A[j]
```
